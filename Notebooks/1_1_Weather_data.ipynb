#WEATHER DATA FRAME

### 1. Importing libraries

```python
# importing libraries
import numpy as np
import pandas as pd
import os
```

###2.Importing files

```python
# to import files in google colab
from google.colab import files
upload=files.upload()
```

```python
# Get all files in the /content/ directory
all_files = os.listdir('/content/')

# Filter out files that are not the ones you uploaded and end with .csv
csv_files = [f'/content/{file}' for file in all_files if "Houston,TX" in file and file.endswith('.csv')]

# Sort the list
csv_files.sort()

print(csv_files)
```

### 3. Merging files

```python
# create empty dataframe
full_weather_df = pd.DataFrame()
```

```python
# Create a list of DataFrames
dfs = [pd.read_csv(file) for file in csv_files]

# Concatenate all the DataFrames together
full_weather_df = pd.concat(dfs, ignore_index=True)
```

### 4. Saving and downloading marged file

```python
# save the merged file
full_weather_df.to_csv('full_weather_df.csv', index=False)
```

```python
# Download the CSV
files.download('full_weather_df.csv')
```

### 5. Exploring merged file

```python
weather_df = full_weather_df.copy()
```

```python
weather_df.head(3)
```

```python
weather_df.tail(3)
```

```python
weather_df.info()
```

Dropping column 'severerisk' as data has been collected only from 2022

```python
weather_df = weather_df.drop(columns=['severerisk'])
```

In the column 'preciptype', we had null values when there was no precipitation on the said day, so I replaced null valeus with string 'none'

```python
weather_df['preciptype'].fillna('none', inplace=True)
```

In the column 'windgust', we had null values on days that no windgust has been reported, therefor I replaced null valeus with the 0

```python
weather_df['windgust'].fillna(0, inplace=True)
```

For some reason 'solarradiation', 'solarenergy', 'uvindex' have not been recorded for 2014 year in visula crossing data, we can exclude either this year or this features, but for now I left them the way they are.

I chanded data type in 'datetime', 'sunrise' and 'sunset' to datetime

```python
weather_df['datetime'] = pd.to_datetime(weather_df['datetime'], format='%d/%m/%Y')
```

first needed to reformat sunrise and sunset columns

```python
weather_df['sunrise'] = weather_df['sunrise'].str.replace('T', ' ')
weather_df['sunset'] = weather_df['sunset'].str.replace('T', ' ')
```

data type can be changed

```python
weather_df['sunrise'] = pd.to_datetime(weather_df['sunrise'], format='%Y-%m-%d %H:%M:%S')
weather_df['sunset'] = pd.to_datetime(weather_df['sunset'], format='%Y-%m-%d %H:%M:%S')
```
