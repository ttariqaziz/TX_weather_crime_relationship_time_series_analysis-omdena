## Merge Offence_type columns into one

```python
# function to merge 'Offence_type' columns conditionally
def merge_offense_type(row):
    if pd.notna(row['Offense_type']):
        return row['Offense_type']
    elif pd.notna(row['Offence_type']):
        return row['Offence_type']
    else:
        return None

# create a new merged column 'Merged_Offense_Type'
final_crime_file['Merged_Offense_Type'] = final_crime_file.apply(merge_offense_type, axis=1)

# drop the two 'Offense_type' columns
final_crime_file.drop(['Offense_type', 'Offence_type'], axis=1, inplace=True)
```

```python
final_crime_file.head(2)
```

```python
final_crime_file.info()
```

## Merge Street_type columns into one

```python
# function to merge 'Street_type' columns conditionally
def merge_street_types(row):
    if pd.notna(row['Street_type']):
        return row['Street_type']
    elif pd.notna(row['StreetType']):
        return row['StreetType']
    else:
        return None

# create a new merged column 'Merged_Offense_Type'
final_crime_file['Merged_Street_Type'] = final_crime_file.apply(merge_street_types, axis=1)

# drop the two 'Offense_type' columns
final_crime_file.drop(['Street_type', 'StreetType'], axis=1, inplace=True)
```
